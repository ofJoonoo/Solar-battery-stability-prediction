```python
# categorical columns
# 최종적으로 변수를 n 개만 남기는 전처리
# 1~n-1 개 까지 범주만 남기고 나머지는 etc 로 value 변경
def top_n_value_with_etc(df, col, n):
    n = int(n)
    survive_value_list = df[col].value_counts().index[:n] # 상위 n-1개 범주 list 생성
    df[col] = np.where(df[col].isin(survive_value_list), df[col], 'etc') # 상위 범주에 속하면 그대로, 아니면 'etc' 로 변경
    return df[col]
```

```python
# 상위 4개만 유지(2개)
val_4 = ["HTL_deposition_procedure", "JV_test_atmosphere"]
# 상위 5개만 유지(3개)
val_5 = ["Perovskite_composition_b_ions_coefficients", "Backcontact_stack_sequence", "Perovskite_deposition_synthesis_atmosphere", ]

val = [val_4, val_5]
for val in val:
    if val == val_4:
        for col in val:
            df[col] = top_n_value_with_etc(df, col, 4)
    elif val == val_5:
        for col in val:
            df[col] = top_n_value_with_etc(df, col, 5)
```
